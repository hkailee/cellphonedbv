```python
import numpy as np
import pandas as pd
from scipy.stats import mstats

def using_mstats_df(df):
    return df.apply(using_mstats, axis=0)

def using_mstats(s):
    return mstats.winsorize(s, limits=[0.05, 0.05])
```

```python
df = pd.read_csv('./in/TPM_EnsemblID_cellphonedb_normal_SC019.txt', sep='\t')
df.set_index('Unnamed: 0', inplace=True)
print(df.loc['ENSG00000001461'].quantile([0.95]))
sorted(df.loc['ENSG00000001461'])
```

```python
def WinsorizeSampleList(data):
    quantiles = data.quantile([0.95])
    # q_05 = quantiles.loc[0.05]
    q_95 = quantiles.loc[0.95]

    # return data[(data.values >= q_05) & (data.values <= q_95)]
    return data[(data.values >= q_95)]
```

```python
def WinsorizeCustom(data):
    quantiles = data.quantile([0.05, 0.95])
    q_05 = quantiles.loc[0.05]
    q_95 = quantiles.loc[0.95]

    out = np.where(data.values <= q_05,q_05, 
                                      np.where(data >= q_95, q_95, data)
                  )
    return out
```

```python
df.apply(WinsorizeCustom, axis=1)
```
